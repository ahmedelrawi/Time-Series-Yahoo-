### DickeyFuller test, Test the accuacy of data related to uniroot square value

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(timeseries):
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
       dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)

print(adf_test(data['Open'].resample('MS').mean()))
```

```python
df_open_t_adj = data['Open'].resample('MS').mean() - data['Open'].resample('MS').mean().shift(1)
df_open_t_adj = df_open_t_adj.dropna()
df_open_t_adj.plot()

print(adf_test(df_open_t_adj))
```

# Data Preprocessing 

```python
# Choose the target column 'High'

trained_data = data['High'].iloc[:-4]
pd.DataFrame(trained_data)
```

```python
# Know the train length

import math
train_len = math.ceil(len(trained_data)*0.8)
train_len
```

```python
# Define a new 2 columns contain the train of x and y with sorting,The following lines create the training sets
# the first row takes the first 10 time steps,
# and the second row takes time steps shifted in one and so on and so forth 


X_train=[]
Y_train=[]

for i in range(6, len(trained_data)):
    X_train.append(trained_data[i-6:i])
    Y_train.append(trained_data[i])
```

```python
# translate it into numpy array
X_train, Y_train= np.array(X_train), np.array(Y_train)
```

```python
# Reshaping for the model
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_train.shape
```

# Deep Learning Model

```python
# For Model and apply RNN + LSTM
from keras.layers import LSTM
from keras.models import Sequential
from keras.layers.wrappers import TimeDistributed
from keras.layers.core import Dense, Activation, Dropout
```

```python
# make the model
model=Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1],1)))
model.add(Dense(25))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
model.fit(X_train, Y_train, epochs=10, batch_size=10, verbose=0)
```
